### **HE vs CLAHE**

```python
import cv2
```

```python
def compare_he_clahe(image_path, clip_limit=2.0, tile_grid=(8, 8), show=True):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    he = cv2.equalizeHist(img)

    clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=tile_grid)
    clahe_img = clahe.apply(img)

    if show:
        import matplotlib.pyplot as plt

        plt.figure(figsize=(12,4))
        plt.subplot(1,3,1)
        plt.title("Original")
        plt.imshow(img, cmap="gray")
        plt.axis("off")

        plt.subplot(1,3,2)
        plt.title("HE")
        plt.imshow(he, cmap="gray")
        plt.axis("off")

        plt.subplot(1,3,3)
        plt.title("CLAHE")
        plt.imshow(clahe_img, cmap="gray")
        plt.axis("off")

        plt.show()

```

```python
compare_he_clahe("/content/fractured arm.jpg")
```
